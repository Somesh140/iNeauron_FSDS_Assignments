```python
def read_replace(txt_file_name,original_string,new_string):
    """function to read text file and replace specific content of the file 
    
   Args:
   
   txt_file_name(str): location of txt file
    
    original_string(str) : string that is to be replaced with new string
    
    new_string(str) : new string  """
    try:
        with open(txt_file_name,"r+") as file:
       #read the contents of the file
            content = file.read()
       #search the string 
            if original_string in content:
            #replace the original string with new string
                content = content.replace(original_string,new_string)
            #delete the old content
                file.seek(0)
                file.truncate()
            #write the new content  
                file.write(content)
                print("{} replaced with {}".format(original_string,new_string))
            else:
                print("{} not found in file".format(original_string))
    except Exception as e:
        print("Error : ",e)
```

```python
#replacing placement in example.txt with screening
read_replace("example.txt","placement","screening")
```

## Demonstration for abstract class

```python
#Abstract Class
from abc import ABC,abstractmethod

class Shape(ABC):#base class
    @abstractmethod
    def printarea(self):
        return 0
    
class circle(Shape):
    type="circle"
    def __init__(self):
        self.radius= 10
    # since we have defined printarea as an abstract method we need to compulsorily define it in child class  
    def printarea(self):
        return 3.14*self.radius**2
```

```python
circle_1 = circle()
print(circle_1.printarea())
```

## Demonstration for Multiple inheritence example

```python
#Multiple Class inheritence
class country:#parent class 1
    country_name = "India"
    continent = 'Asia'
 
    def location(self):
        print('Present  location is  ', self.country_name)

class state:#parent class 2
    state_name= "Karnataka"
    region = "Peninsula"
 
 
class city(country,state):#child class inherited attributes from both parent class
    city_name="Bangalore"
    def __init__(self):
        if self.state_name == "Karnataka":
            print('Welcome to Karnataka (info inherited from state class)')
        print('Country: {} (info inherited from country class)'.format(self.country_name))
```

```python
#object of class city
city_1=city()
```

## Demonstration for decorator example

```python
#use of decorator

def print_before_after(func):
    """decorator that prints "Before" message before decorated function and "After" message after decorated function """ 
    def wrapper(*args):
        print(" Before {} ".format(func.__name__))
        #call the function being decorated with arguments
        func(*args)
        print("After {} ".format(func.__name__))
        #return the nested function 
    return wrapper 
    
@print_before_after    
def multiply(a,b):
    """function that multiplies the 2 arguments passsed
    Args:
    a:int
    b:int """
    print(a*b)
    

multiply(5,10)
```

```python
#without decorator
def multiply(a,b):
    """function that multiplies the 2 arguments passsed
    Args:
    a:int
    b:int """
    print(a*b)
    
multiply(5,10)
```
