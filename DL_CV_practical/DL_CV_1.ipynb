Objective: - Predict Next Sequence
To start with deep learning, the very basic project that you can build is to
predict the next digit in a sequence.

Dataset: - Create a sequence like a list of odd numbers and then build a model
and train it to predict the next digit in the sequence.

Task: - A simple neural network with 2 layers would be sufficient to build the
model.

```python
#We can define a function to generate the sequence up to a specified length.
def generate_sequence(length):
    sequence = []
    num = 1
    for _ in range(length):
        sequence.append(num)
        num += 2
    return sequence

sequence_length = 100
sequence = generate_sequence(sequence_length)
```

```python
sequence
```

we're using a simple feedforward neural network with one input neuron and one output neuron. The model has two layers: a hidden layer with 10 neurons and a linear output layer. We compile the model using the mean squared error loss and the Adam optimizer.

To train the model, we use the generated sequence as both the input (X) and the target output (y). We're essentially training the model to predict the next digit in the sequence based on the current digit. We fit the model to the data for 100 epochs with a batch size of 1.

After training, you can use the model to predict the next digit in the sequence by providing the last digit as input.

```python
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense

# Define the model
model = Sequential()
model.add(Dense(10, input_dim=1, activation='relu'))
model.add(Dense(1, activation='linear'))

# Compile the model
model.compile(loss='mean_squared_error', optimizer='adam')

# Train the model
X = sequence[:-1]
y = sequence[1:]
model.fit(X, y, epochs=100, batch_size=1, verbose=2)
```

```python
last_digit = sequence[-1]
next_digit = model.predict([last_digit])
print("Predicted next digit:", next_digit)
```
